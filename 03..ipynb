## 3. Python program to display IMDBâ€™s Top rated 100 Indian movies

```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://www.imdb.com/list/ls056092300/'
page=requests.get(url)
page
```

```python
soup = BeautifulSoup(page.content)
Name=[]
Rating=[]
Year=[]

for i in soup.find_all('div',class_='lister-item-content'):
    Name.append(i.find('a').text)
Name
```

```python
for i in soup.find_all('div',class_='ipl-rating-star small'):
    Rating.append(i.find('span',class_='ipl-rating-star__rating').text)
Rating
```

```python
for i in soup.find_all('h3',class_='lister-item-header'):
    Year.append(i.find('span',class_='lister-item-year text-muted unbold').text)
Year
```

```python
import pandas as pd
df=pd.DataFrame({'Name':Name,'Rating':Rating,'Year':Year})
df
```
