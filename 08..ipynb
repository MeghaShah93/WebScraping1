```python
soup=BeautifulSoup(page.content)
```

```python
housetitle=[]
for i in soup.find_all('h2', class_='heading-6 font-semi-bold nb__25Cl7'):
    housetitle.append(i.text)
housetitle
```

```python
location=[]
for i in soup.find_all('div',class_='nb__1EwQz'):
    location.append(i.text)
location
```

```python
areas = []
EMI = []
prices = []

for i in soup.find_all('div', class_='nb__3q08s'):
    EMI.append(i.find_all('div',class_='font-semi-bold heading-6')[1].text)
    prices.append(i.find_all('div',class_='nb__4L90a')[3].find('div', class_='font-semi-bold heading-6').text.replace('₹',''))
    areas.append(i.find_all('div',class_='nb__4L90a')[1].find('div',class_='nb__FfHqA').text)
print(areas)
print(EMI)
print(prices)
```

```python
import pandas as pd
df = pd.DataFrame({'Housetitle':housetitle,'Location':location,'Area':areas,'EMI':EMI,'Price':prices})
df
```
