### 2. Python program to display IMDB’s Top rated 100 movies

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc'
url2 = 'https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt'
page = requests.get(url)
page2 = requests.get(url2)
print(page)
print(page2)
```

```python
soup = BeautifulSoup(page.content)
soup2 = BeautifulSoup(page2.content)
Name = []
Rating = []
#Year = []

for i in soup.find_all('div',class_='lister-item-content'):
    Name.append(i.find('a').text)
for i in soup2.find_all('div',class_='lister-item-content'):
    Name.append(i.find('a').text)

Name
```

```python
for i in soup.find_all('div',class_='ratings-bar'):
    Rating.append(i.find('strong').text)
for i in soup2.find_all('div',class_='ratings-bar'):
    Rating.append(i.find('strong').text)
Rating
```

```python
Years = []
for i in soup.find_all('div',class_='lister-item-content'):
    Years.append (i.find('span',class_='lister-item-year text-muted unbold').text)
for i in soup2.find_all('div',class_='lister-item-content'):
    Years.append (i.find('span',class_='lister-item-year text-muted unbold').text)
Years
```

```python
import pandas as pd
df=pd.DataFrame({'Name':Name,'Rating':Rating,'Year':Years})
df
```
